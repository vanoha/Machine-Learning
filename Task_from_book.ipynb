```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
movies = ["Andy Holl", "Ban-Gur", "Casablanca", "Gandy", "Westside story"]
num_oscars = [5, 11, 3, 8, 10]

xs = [i + 0.1 for i, _ in enumerate(movies)]

plt.bar(xs, num_oscars)

plt.ylabel("Amount of awards")
plt.title("Favorite films")

plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
hist = Counter(decile(grade) for grade in grades)

plt.bar(hist.keys(), hist.values(), 8)
plt.axis([-5, 105, 0, 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("Amount of students")
plt.title("Exam rating distribution № 1")
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label = 'dispersion')
plt.plot(xs, bias_squared, 'r--', label = 'bias')
plt.plot(xs, total_error, 'b:', label = 'total error')

plt.legend(loc=9)
plt.xlabel('Complexity of model')
plt.title('Compromise between bias and dispersion')
plt.show()
```

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
  plt.annotate(label, 
               xy=(friend_count, minute_count),
               xytext=(5, -5),
               textcoords='offset points')

plt.title("Dependence between minutes and friends")
plt.xlabel("Friends")
plt.ylabel("Minutes online in site")
plt.axis('equal')
plt.show()
```
